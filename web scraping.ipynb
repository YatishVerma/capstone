```python
import selenium
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.action_chains import ActionChains
```

```python
cities=['Austin TX',"Buffalo NY","Dallas","New Jersey,USA","Las Vegas,NV","Miami,FL","Orlando,FL,USA"
        ,"Ottawa,CA","Vancouver, CA ","Toronto, CA ","Brisbane","Melbourne","Sydney","Tokyo","Perth",
       "Amsterdam","Birmingham","Glasgow","Frankfurt","Munich"]
pageiter=[8,8,9,10,10,9,9,9,9,9,9,9]
```

```python
options = webdriver.ChromeOptions()
options.add_argument('--headless')

driver = webdriver.Chrome(options=options)
new_options = webdriver.ChromeOptions()

# Initialize the WebDriver without headless mode
driver = webdriver.Chrome(options=new_options)
```

```python
driver.get("https://rentberry.com")
```

```python
City=[]
Price=[]
Rooms=[]
Bath=[]
Area=[]
Location=[]
```

```python
for city in cities:
    time.sleep(5)
    xpath='/html/body/rb-core/div/rb-home-page/rb-promo-search/div/div/div/div/div[2]/input'
    element=driver.find_element(By.XPATH,xpath)
    element.send_keys(city)
    #need some time delay here
    time.sleep(3)
    element.send_keys(Keys.ENTER)
    time.sleep(5)
    driver.maximize_window()
    time.sleep(5)
    xpath='//*[@id="search-sidebar-wrapper"]/div[1]/div[1]/div[2]/button/span[2]/span'
    hide_map=driver.find_element(By.XPATH,xpath)
    hide_map.click()
    time.sleep(5)
    scroll=driver.find_element(By.XPATH,'/html/body')
    scroll.click()
    for x in range(1,7,1):
        scroll.send_keys(Keys.PAGE_DOWN)
    for x in range(1,25,1):
        try:
            price=driver.find_element(By.XPATH,'//*[@id="search-sidebar-wrapper"]/div[1]/rb-filter-bar/rb-sidebar/div/div[1]/div[{}]/rb-property-card/section/rb-property-card-bottom/div/div/rb-property-card-price/div/p'.format(x))
            pric=price.text
            City.append(city)
            Price.append(pric)
               
        except:
            Price.append("")
            City.append("")
        try:
            room=driver.find_element(By.XPATH,'//*[@id="search-sidebar-wrapper"]/div[1]/rb-filter-bar/rb-sidebar/div/div[1]/div[{}]/rb-property-card/section/rb-property-card-bottom/div/rb-property-card-labels/div/div[1]'.format(x))
            rooms=room.text
            Rooms.append(rooms)
        
        except:
            Rooms.append("")
        try:
            bath=driver.find_element(By.XPATH,'//*[@id="search-sidebar-wrapper"]/div[1]/rb-filter-bar/rb-sidebar/div/div[1]/div[{}]/rb-property-card/section/rb-property-card-bottom/div/rb-property-card-labels/div/div[2]'.format(x))
            baths=bath.text
            Bath.append(baths) 
        except:
            Bath.append("")
        try:
            area=driver.find_element(By.XPATH,'//*[@id="search-sidebar-wrapper"]/div[1]/rb-filter-bar/rb-sidebar/div/div[1]/div[{}]/rb-property-card/section/rb-property-card-bottom/div/rb-property-card-labels/div/div[3]'.format(x))
            areas=area.text
            Area.append(areas)
        
        except:
            Area.append("")
                #2 is meter square's 2
        try:
            loc=driver.find_element(By.XPATH,'//*[@id="search-sidebar-wrapper"]/div[1]/rb-filter-bar/rb-sidebar/div/div[1]/div[{}]/rb-property-card/section/rb-property-card-bottom/div/rb-property-card-address/div/a'.format(x))
            location=loc.text
            Location.append(location)
        
        except:
            Location.append("")
                        
            
    time.sleep(5)
    for k in pageiter:
        try:
            element=driver.find_element(By.XPATH,'//*[@id="search-sidebar-wrapper"]/div[1]/rb-filter-bar/rb-sidebar/div/div[2]/rb-search-pagination/div[2]/a[{}]/span'.format(k))
#             element=driver.find_element(By.XPATH,'//*[@id="search-sidebar-wrapper"]/div[1]/rb-filter-bar/rb-sidebar/div/div[2]/rb-search-pagination/div[2]/a[3]')
            element.click()
            time.sleep(5)
            print("--"*5)
        except:
            print((f"Element not found for page {k}"))
        for x in range(1,25,1):
            try:
                price=driver.find_element(By.XPATH,'//*[@id="search-sidebar-wrapper"]/div[1]/rb-filter-bar/rb-sidebar/div/div[1]/div[{}]/rb-property-card/section/rb-property-card-bottom/div/div/rb-property-card-price/div/p'.format(x))
                pric=price.text
                City.append(city)
                Price.append(pric)
               
            except:
                Price.append("")
                City.append("")
            try:
                room=driver.find_element(By.XPATH,'//*[@id="search-sidebar-wrapper"]/div[1]/rb-filter-bar/rb-sidebar/div/div[1]/div[{}]/rb-property-card/section/rb-property-card-bottom/div/rb-property-card-labels/div/div[1]'.format(x))
                rooms=room.text
                Rooms.append(rooms)
        
            except:
                Rooms.append("")
            try:
                bath=driver.find_element(By.XPATH,'//*[@id="search-sidebar-wrapper"]/div[1]/rb-filter-bar/rb-sidebar/div/div[1]/div[{}]/rb-property-card/section/rb-property-card-bottom/div/rb-property-card-labels/div/div[2]'.format(x))
                baths=bath.text
                Bath.append(baths) 
            except:
                Bath.append("")
            try:
                area=driver.find_element(By.XPATH,'//*[@id="search-sidebar-wrapper"]/div[1]/rb-filter-bar/rb-sidebar/div/div[1]/div[{}]/rb-property-card/section/rb-property-card-bottom/div/rb-property-card-labels/div/div[3]'.format(x))
                areas=area.text
                Area.append(areas)
        
            except:
                Area.append("")
                #2 is meter square's 2
            try:
                loc=driver.find_element(By.XPATH,'//*[@id="search-sidebar-wrapper"]/div[1]/rb-filter-bar/rb-sidebar/div/div[1]/div[{}]/rb-property-card/section/rb-property-card-bottom/div/rb-property-card-address/div/a'.format(x))
                location=loc.text
                Location.append(location)
        
            except:
                Location.append("")
       
    driver.get("https://rentberry.com")
    
    
    
```
