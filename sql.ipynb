```python
#   Write a SQL query to Select records where both Washer/Dryer and AC are available, and order by Sno.
qry="""SELECT * FROM amenitiestable WHERE `Air Conditioning` = 1 AND `Washer/Dryer In Unit` =1 ORDER BY sno """
cursor.execute(qry)
result = cursor.fetchall()
result
```

```python
#Write a SQL query to Select records where Hardwood floors are
# available but neither Roofdeck nor Storage is present, and order by Sno in descending order.
qry="""SELECT * FROM amenitiestable WHERE `Roof Deck` = 0 AND `Hardwood Floors` = 1 AND Storage = 0 ORDER BY sno DESC"""
cursor.execute(qry)
result = cursor.fetchall()
result
```

```python
# Write a SQL query to Select records where at least four 
#amenities (AC, Parking, Dishwasher, Fireplace) are available, and order by Sno .
qry="""SELECT *
FROM amenitiestable
WHERE  `Air Conditioning`= 1 AND Parking = 1 AND Dishwasher = 1 AND Fireplace = 1
ORDER BY Sno;"""
cursor.execute(qry)
result = cursor.fetchall()
result
```

```python
#  Write a SQL query to Select records where neither Roofdeck nor Storage is available, and include the count of such records.
qry="""SELECT COUNT(*) AS RecordCount
FROM amenitiestable
WHERE `Roof Deck` = 0 AND Storage = 0"""
cursor.execute(qry)
result = cursor.fetchall()
result
```

```python
#  Write a SQL query to Select records with Parking and
#either Fireplace or Dishwasher, and include the count of records for each condition.
qry="""SELECT Parking,Fireplace,Dishwasher, COUNT(*) AS recordcount FROM amenitiestable WHERE Parking =1 AND (Fireplace = 1 OR Dishwasher =1) GROUP BY Fireplace,Dishwasher,Parking"""
cursor.execute(qry)
result = cursor.fetchall()
result
```

SUB QUERIES

```python
#Write a SQL  subquery to find records with more than the average area and related details using table 1 and table 2.
qry="""SELECT * FROM pricetable t1 JOIN roomstable t2 ON t1.sno = t2.sno WHERE t2.Converted_Area > (SELECT AVG(Converted_Area) FROM roomstable)"""
cursor.execute(qry)
result = cursor.fetchall()
result
```

```python
#Write a subquery to find records in table1 based on conditions pets allowed is YES  and no of bed is greater than 3   in table2.
qry="""SELECT * FROM roomstable t1 JOIN pricetable t2 ON t1.sno=t2.sno WHERE t1.Dogs_allowed = 'allowed' AND t1.Rooms > 3 """
cursor.execute(qry)
result = cursor.fetchall()
result
```

```python
# Write a SQL    subquery using both tables (2 and 3) to find records in
# Table2 with more than 2 bedrooms and related details from Table3 where AC is present .
qry="""SELECT * FROM roomstable t1 JOIN amenitiestable t2 ON t1.sno=t2.sno WHERE t2.`Air Conditioning` = 1 AND t1.Rooms > 2 """
cursor.execute(qry)
result = cursor.fetchall()
result
```

```python
#Write a sql subquery  to find records in Table2 with pets allowed and a Dishwasher, and include related details from Table3.
qry="""SELECT * FROM roomstable t1 JOIN amenitiestable t2 ON t1.sno=t2.sno WHERE t2.Dishwasher = 1 AND t1.Dogs_allowed = 'allowed' """
cursor.execute(qry)
result = cursor.fetchall()
result
```

```python
#Write a subquery  to find records in Table2 with the highest area and related details from Table3 where roofdeck is present.
qry="""SELECT * FROM roomstable t1 JOIN amenitiestable t2 ON t1.sno=t2.sno WHERE t2.`Roof Deck`= 1 ORDER BY Converted_Area DESC"""
cursor.execute(qry)
result = cursor.fetchall()
result
```

```python
#     Write a sql  Inner Join to combine information from table1 and table 2.
qry="""SELECT *
FROM pricetable
INNER JOIN roomstable ON pricetable.sno = roomstable.sno"""
cursor.execute(qry)
result = cursor.fetchall()
result
```

```python
# Write SQL  Subquery to find records in table1 with pets allowed and a Washer/Dryer, and include details from table2 and table3 
qry="""SELECT t1.*, t2.*, t3.*
FROM pricetable t1
JOIN roomstable t2 ON t1.sno = t2.sno
JOIN amenitiestable t3 ON t1.sno = t3.sno
WHERE t2.Dogs_allowed = 'allowed' AND t3.`Washer/Dryer In Unit` = 1;"""
cursor.execute(qry)
result = cursor.fetchall()
result
```
